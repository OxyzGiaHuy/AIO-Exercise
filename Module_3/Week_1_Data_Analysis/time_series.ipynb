1. Import libraries and read dataset

```python
import pandas as pd

dataset_path = 'dataset/opsd_germany_daily.csv'
# Open Power System Data
opsd_daily = pd.read_csv(dataset_path)
opsd_daily
```

```python
opsd_daily = opsd_daily.set_index('Date')
opsd_daily
```

```python
opsd_daily = pd.read_csv(dataset_path, index_col=0, parse_dates=True)

# Add columns with year , month , and weekday name
opsd_daily['Year'] = opsd_daily.index.year
opsd_daily['Month'] = opsd_daily.index.month
opsd_daily['Weekday Name'] = opsd_daily.index.day_name()

# Display a random sampling of 5 rows
opsd_daily.sample(5, random_state=0)
```

3. Visualizing time series data

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(rc={'figure.figsize': (15,4)}) # rc: runtime config, allow seaborn use config of matplotilb to visualize
opsd_daily['Consumption'].plot(linewidth=0.5)
```

```python
# load multiple columns
col_plot = ['Consumption', 'Solar', 'Wind']
axes = opsd_daily[col_plot].plot(marker = '.', alpha = .5, linestyle = 'None', figsize=(15,9), subplots=True)
for ax in axes:
    ax.set_ylabel('Daily Total (GWh)')
plt.show()
```

4. Seasonality

```python
fig, axes = plt.subplots(3, 1, figsize = (15, 6), sharex=True)
for name, ax in zip(col_plot, axes):
    sns.boxplot(data=opsd_daily, x='Month', y = name, ax = ax, hue = 'Month', palette = sns.color_palette("Set3", 12), legend=False)
    # hue param to choose variable to cluster color, palette is palette, legend to turn off annotation
    ax.set_ylabel('GWh')
    ax.set_title(name)
```

```python
times_sample = pd.to_datetime(['2013-02-03', '2013-02-06', '2013-02-08'])
consumpt_sample = opsd_daily.loc[times_sample, ['Consumption']].copy()
consumpt_freq = consumpt_sample.asfreq('D')
consumpt_freq['Consumption - Forward Fill'] = consumpt_sample.asfreq('D', method='ffill') # forward fill: fill missing values by previous row data
consumpt_freq
```

6. Resampling

```python
# downsampling
data_columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']
opsd_weekly_mean = opsd_daily[data_columns].resample('W').mean() # must include a method (at this point is mean) as Groupby 
opsd_weekly_mean
```

```python
start, end = '2017-01', '2017-06'
plt.figure(figsize=(15,9))
plt.plot(opsd_daily.loc[start:end, 'Solar'], marker = '.', linewidth = 0.5, label = 'Daily')
plt.plot(opsd_weekly_mean.loc[start:end, 'Solar'], marker = 'o', markersize = 8, label = 'Weekly Mean Resample')
plt.ylabel('Solar Production (GWh)')
plt.legend()
plt.show()
```

```python
start, end = '2017-01', '2017-06'
plt.figure(figsize=(15,9))
plt.plot(opsd_daily.loc[start:end, 'Wind'], marker = '.', linewidth = 0.5, label = 'Daily')
plt.plot(opsd_weekly_mean.loc[start:end, 'Wind'], marker = 'o', markersize = 8, label = 'Weekly Mean Resample')
plt.ylabel('Wind Production (GWh)')
plt.legend()
plt.show()
```

```python
opsd_annual = opsd_daily[data_columns].resample('YE').sum()
opsd_annual
```

```python
opsd_annual = opsd_daily[data_columns].resample('YE').sum(min_count=360)
opsd_annual
```

```python
opsd_annual = opsd_annual.set_index(opsd_annual.index.year)
opsd_annual
```

```python
opsd_annual.index.name = "Year"
opsd_annual
```

```python
opsd_annual['Wind+Solar/Consumption'] = opsd_annual['Wind+Solar']/opsd_annual['Consumption']
opsd_annual
```

```python
# because there is no solar production data in years that are earlier than 2012:
plt.figure(figsize=(11,4))
ax = opsd_annual.loc[2012:, 'Wind+Solar/Consumption'].plot.bar()
ax.set_ylabel('Fraction')
ax.set_ylim(0, 0.3)
ax.set_title('Wind + Solar Share of Annual Electricity Consumption')
plt.xticks(rotation = 0)
```

7. Rolling windows

```python
opsd_7d = opsd_daily[data_columns].rolling(7, center=True).mean()
opsd_7d
# 2006-01-04 represents the 7-day period from 2006-01-01 to 2006-01-07
```

8. Trend

```python
import matplotlib.dates as mdates
opsd_365d = opsd_daily[data_columns].rolling(window=365, center = True, min_periods=360).mean()
opsd_365d
```

```python
fig, ax = plt.subplots(figsize=(11,4))
ax.plot(opsd_daily['Consumption'], marker = '.', markersize =2, color = '0.6', linestyle = 'None', label = 'Daily')
ax.plot(opsd_7d['Consumption'], linewidth = 2, label = '7-d Rolling Mean')
ax.plot(opsd_365d['Consumption'], color = '0.2', linewidth = 3, label = 'Trend (365-d Rolling Mean)')
# Set x-ticks to yearly interval and add legend and labels
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Consumption (GWh)')
ax.set_title('Trends in Electricity Consumption')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (11,4))
for name in ['Wind', 'Solar', 'Wind+Solar']:
    ax.plot(opsd_365d[name], label = name)
    # Set x- ticks to yearly interval , adjust y- axis limits , add legend and labels
    ax.xaxis.set_major_locator(mdates.YearLocator())
ax.set_ylim(0, 400)
ax.set_ylabel('Production (GWh)')
ax.set_title('Trends in Electricity Production (365-d Rolling Means)')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (11,4))
for name in ['Wind', 'Solar', 'Wind+Solar']:
    ax.plot(opsd_365d[name], label = name)
ax.set_ylim(0, 400)
ax.set_ylabel('Production (GWh)')
ax.set_title('Trends in Electricity Production (365-d Rolling Means)')
plt.show()
```
